```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans

# Load the dataset
df = pd.read_csv('olympics2024.csv')

# Select features for clustering
X = df[['Gold', 'Total']]

# Standardize the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply K-means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)

# View the result
print(df[['Country', 'Gold', 'Silver', 'Bronze', 'Total', 'Cluster']])

# Plotting the clusters
plt.figure(figsize=(10, 6))
plt.scatter(df['Gold'], df['Total'], c=df['Cluster'], cmap='rainbow', marker='o')
plt.title('K-means Clustering Visualization Based on Gold and Total Medals')
plt.xlabel('Gold Medals')
plt.ylabel('Total Medals')
plt.colorbar(label='Cluster')
plt.show()
```
